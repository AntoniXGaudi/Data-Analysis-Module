```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
labels = ["Africa", "America", "Asia", "Europe"]
values = [10, 5, 20, 15]
```

```
sns.barplot(x=labels, y=values);
```

```
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = labels
sizes = values
explode = ( 0, 0, 0, 0.2)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots(figsize=(7,7))

ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=110)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df.origin.value_counts().index
sizes = df.origin.value_counts().values
explode = ( 0, 0.2, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig, ax = plt.subplots(figsize=(7,7))

ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=110)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```
import numpy as np
from scipy.stats import kurtosis, skew
```

```
from scipy import stats
```

```
x=np.random.normal(0,2,10000000)
plt.hist(x,bins=100);
```

```
kurtosis(x)
```

```
skew(x)
```

```
shape, scale = 2, 2
s=np.random.gamma(shape,scale, 100000)
```

```
plt.hist(s, bins=100);
```

```
kurtosis(s)
```

```
skew(s)
```
